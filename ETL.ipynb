<h3>Concatenamos las 4 tablas en un dataframe.<h3>

```python
#Comenzamos por ver el número de filas de cada tabla para darnos una idea de cómo puede quedar el dataframe concatenado.

print("Número de filas por tabla","\nAmazon: ",df_amazon.shape[0])
print("Disney: ",df_disney.shape[0])
print("Hulu:   ",df_hulu.shape[0])
print("Netflix:",df_netflix.shape[0])
```

```python
#Insertamos una columna "platform" a cada una de las tablas para saber a qué plataforma refiere la fila.

df_amazon = df_amazon.assign(platform="amazon")
df_disney = df_disney.assign(platform="disney")
df_hulu = df_hulu.assign(platform="hulu")
df_netflix = df_netflix.assign(platform="netflix")
```

```python
df_amazon.head()
```

```python
#Concatenamos verticalmente las 4 tablas en un solo dataframe "df_final" y utilizamos ignore_index para resetear los índices.

df_final = pd.concat([df_amazon,df_disney,df_hulu,df_netflix], ignore_index=True)
df_final.shape
```

```python
#Traemos una sample del df_final para verificar que se haya concatenado de manera correcta.

df_final.head()
```

```python
# Separo mi columna 'duration' en 'duration_int' y 'duration_type'

def sep_duration(df):

    df = df.join(df.duration.str.split(expand=True))

    df.rename(columns={0:'duration_int',
                                1:'duration_type'}, inplace=True)
    return df
```

```python
df_final = sep_duration(df_final)
```

```python
df_final.head()
```

```python
# Separo mi columna 'duration' según valor y tipo

df_final.duration_int = df_final.duration_int.replace('Sin',0)
df_final.duration_type = df_final.duration_type.replace('Dato','Sin Dato')
```

```python
df_final.duration_int = df_final.duration_int.astype('int64')
```

```python
# Normalizo mi columna duration_type convirtiendo 'Season' a 'season' para que mantenga la coherencia con 'min' que se encuentra en minúscula

df_final.duration_type = df_final.duration_type.replace('Season','season')
df_final.duration_type = df_final.duration_type.replace('Seasons','season')
```

<h3>Comenzamos con la creación de funciones para realizar las consultas.<h3>

Función get_max_duration.

```python
df_final.duration_int.dtype
```

```python
def get_max_duration(anio,plataforma,tipo):
    #Convierto anio en int en caso de recibirlo como una string
    anio = int(anio)
    #Creo mis tuplas con mis posibles opciones para verificar que el input se encuentre dentro de las mismas.
    tuple_plataformas = ("amazon","disney","hulu","netflix")
    tuple_tipo = ("min","season")
    #Doy las condiciones que deben ser cumplidas para entrar al análisis, de no ser así, pido que se verifiquen los datos ingresados.
    try:
        if (anio in range(df_final.release_year.min(),df_final.release_year.max()+1)) and (plataforma in (tuple_plataformas)) and (tipo in tuple_tipo):
            #Busco el ID de la película/serie con mayor duración que se encuentre en el Dataframe que me devuelve la Query cuyo año, plataforma y tipo recibo como parámetros.
            id_final = df_final.query(f"release_year == {anio} and platform == '{plataforma}' and duration_type == '{tipo}'").duration_int.idxmax()
            #Utilizo la ID recibida para crear un diccionario con los datos del dataframe en esa ID, luego extraigo los values que me interesan informarle al usuario que realiza la consulta.
            return f'{df_final[id_final:id_final+1].to_dict().get("title").get(id_final)} Tiene la mayor duración: {df_final[id_final:id_final+1].to_dict().get("duration").get(id_final)}'
        else:
            return f"Faltan datos para su consulta o alguno de los parámetros es incorrecto, por favor verifique que los datos ingresados."
    except ValueError:
            return f"La plataforma {plataforma} no posee registros en {anio}"
```

```python
def get_count_platform(plataforma):
    #Creo mi tupla con mis posibles opciones para verificar que el input se encuentre dentro de las mismas.
    tuple_plataformas = ("amazon","disney","hulu","netflix")
    #Doy las condiciones que deben ser cumplidas para entrar al análisis, de no ser así, pido que se verifiquen los datos ingresados.
    if plataforma in (tuple_plataformas):
        #Creo un diccionario con el conteo de los títulos que se encuentran en el Dataframe que me devuelve la Query cuya plataforma recibo como parámetro, agrupado por tipo (Movie/TV Show).
        dict = df_final.query(f"platform == '{plataforma}'").groupby(by=['type']).title.count().to_dict()
        #Extraigo los values que me interesan informarme al usuario que realiza la consulta.
        return f'El conteo de la plataforma {plataforma} es: {dict.get("Movie")} películas y {dict.get("TV Show")} series'
    else:
        return f"Faltan datos para su consulta o alguno de los parámetros es incorrecto, por favor verifique que los datos ingresados."
```

```python
from operator import itemgetter #Importo este módulo ya que me sirve para conseguir el valor máximo de un diccionario.
def get_listedin(genero):
    try:
        #Utilizo un str.contains case_sensitive=False sobre mi columna listed_in, luego agrupo por 'platform' y realizo un conteo de los títulos para convertir estos datos en un diccionario.
        dict_plat = df_final[df_final.listed_in.str.contains(genero, case=False)].groupby(by=['platform']).title.count().to_dict()
        max_value = max(dict_plat.items(), key=itemgetter(1))[0]
        #Extraigo los values que me interesan informarme al usuario que realiza la consulta.
        return f'la plataforma {max_value} tiene la mayor frecuencia de género {genero}, apareciendo {dict_plat.get(max_value)} veces'
    except ValueError:
        return f'No hay registros de {genero} en ninguna de nuestras plataformas.'
```

```python
def get_actor(plataforma, anio):
    #Convierto anio en int en caso de recibirlo como una string
    anio = int(anio)
    #Creo mi tupla con mis posibles opciones para verificar que el input se encuentre dentro de las mismas.
    tuple_plataformas = ("amazon","disney","hulu","netflix")
    #Doy las condiciones que deben ser cumplidas para entrar al análisis, de no ser así, pido que se verifiquen los datos ingresados.
    try:
        if (anio in range(df_final.release_year.min(),df_final.release_year.max()+1)) and (plataforma in (tuple_plataformas)):
            #Creo mi lista de actores final vacía
            actor_list = []
            #Creo una lista del cast por película/TV_Show que se encuentren en el Dataframe que me devuelve la Query cuyo año y plataforma recibo como parámetros.
            cast_list = df_final.query(f"platform == '{plataforma}' and release_year == {anio}").cast.tolist()
            #Divido los actores del Cast y los agrego a una lista temporal de actores a la que luego le excluyo los valores 'Sin Datos' para mi lista de actores final.
            for i in range(len(cast_list)):
                actor_list_temp = cast_list[i].split(",")
                for j in range(len(actor_list_temp)):
                    if actor_list_temp[j] != 'Sin Datos':
                        actor_list.append(actor_list_temp[j])
            #Convierto en diccionario mi lista de actores pero únicamente con el valor del que más apariciones tiene.
            dict_actor = dict(zip(actor_list,map(lambda x: actor_list.count(x),actor_list)))
            #Consigo la Key del diccionario que corresponde al Actor/actriz con más apariciones.
            max_actor = max(dict_actor, key=dict_actor.get)
            #Consigo el Value de apariciones de dicho Actor/actriz
            max_actor_appearances = dict_actor.get(max_actor)
            #Extraigo los values que me interesan informarme al usuario que realiza la consulta.
            return f'El actor/actriz con mayor número de apariciones en la plataforma {plataforma} el año {anio} es {max_actor} con {max_actor_appearances} apariciones.'
        else:
            return f"Faltan datos para su consulta o alguno de los parámetros es incorrecto, por favor verifique que los datos ingresados."
    except ValueError:
        return f"La plataforma indicada no posee registros en {anio}"
```
