<h3>Concatenamos las 4 tablas en un dataframe.<h3>

```python
#Comenzamos por ver el número de filas de cada tabla para darnos una idea de cómo puede quedar el dataframe concatenado.

print("Número de filas por tabla","\nAmazon: ",df_amazon.shape[0])
print("Disney: ",df_disney.shape[0])
print("Hulu:   ",df_hulu.shape[0])
print("Netflix:",df_netflix.shape[0])
```

```python
#Insertamos una columna "platform" a cada una de las tablas para saber a qué plataforma refiere la fila.

df_amazon = df_amazon.assign(platform="amazon")
df_disney = df_disney.assign(platform="disney")
df_hulu = df_hulu.assign(platform="hulu")
df_netflix = df_netflix.assign(platform="netflix")
```

```python
df_amazon.head()
```

```python
#Concatenamos verticalmente las 4 tablas en un solo dataframe "df_final" y utilizamos ignore_index para resetear los índices.

df_final = pd.concat([df_amazon,df_disney,df_hulu,df_netflix], ignore_index=True)
df_final.shape
```

```python
#Traemos una sample del df_final para verificar que se haya concatenado de manera correcta.

df_final.head()
```
